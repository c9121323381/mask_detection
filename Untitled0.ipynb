```python
import os
base_dir = "/content/drive/MyDrive/mask_project"
data_dir = os.path.join(base_dir, "mask_dataset")

print("訓練資料路徑：", data_dir)
```

```python
import torch
from torch import nn, optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
# 圖像轉換（資料前處理）
transform = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ])
}
# 載入資料
train_dataset = datasets.ImageFolder(os.path.join(data_dir, 'train'), transform=transform['train'])
val_dataset = datasets.ImageFolder(os.path.join(data_dir, 'val'), transform=transform['val'])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

class_names = train_dataset.classes
print("分類類別:", class_names)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("使用裝置:", device)

# 使用預訓練的 ResNet18
model = models.resnet18(pretrained=True)

# 修改最後一層全連接層（2 類）
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 2)

model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)
```

```python
num_epochs = 10

train_losses, val_losses = [], []

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
    # 驗證
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    train_losses.append(running_loss / len(train_loader))
    val_losses.append(val_loss / len(val_loader))

    print(f"Epoch [{epoch+1}/{num_epochs}]  訓練損失: {running_loss/len(train_loader):.4f}  驗證損失: {val_loss/len(val_loader):.4f}  準確率: {100*correct/total:.2f}%")
```

```python
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.legend()
plt.title('Training & Validation Loss')
plt.show()
```

```python
from PIL import Image

# 測試圖片路徑（你可以換成自己照片）
test_path = os.path.join(data_dir, "val", "with_mask")
test_image = Image.open(os.path.join(test_path, os.listdir(test_path)[0]))

# 預處理
img = transform['val'](test_image).unsqueeze(0).to(device)
model.eval()
with torch.no_grad():
    output = model(img)
    _, pred = torch.max(output, 1)
    pred_class = class_names[pred.item()]

plt.imshow(test_image)
plt.title(f"預測結果：{pred_class}")
plt.axis("off")
plt.show()
```

```python
import random
from PIL import Image
import matplotlib.pyplot as plt

# 隨機選一類（戴或沒戴）
cls = random.choice(class_names)
test_path = os.path.join(data_dir, "val", cls)

# 隨機挑五張圖片
img_names = random.sample(os.listdir(test_path), 5)

plt.figure(figsize=(15,6))

for i, img_name in enumerate(img_names):
    test_image = Image.open(os.path.join(test_path, img_name))

    # 預處理
    img = transform['val'](test_image).unsqueeze(0).to(device)

    # 預測
    model.eval()
    with torch.no_grad():
        output = model(img)
        _, pred = torch.max(output, 1)
        pred_class = class_names[pred.item()]

    # 畫圖
    plt.subplot(1,5,i+1)
    plt.imshow(test_image)
    plt.title(f"真實:{cls}\n預測:{pred_class}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
